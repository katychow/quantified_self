````python
# Want to fill in all missing dates for reasonable time series

full_header = indexes.merge(header, how='left', left_on = 'dates', right_on = 'dateOfSleep')
````

````python
train=full_header_clean[:90]
train.shape
````

````python
test = full_header_clean[90:]
test.head(1),test.tail(1)
````

````python
from statsmodels.tsa.arima_model import ARIMA
````

````python
ar_eff = ARIMA(endog = train['efficiency'], order = (1,0,0))
model = ar_eff.fit()
preds = model.predict(start = 90, end = 120)
mean_squared_error(test['efficiency'],preds)
````

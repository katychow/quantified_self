```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error ,mean_absolute_error
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX

%matplotlib inline
```

```python
df = pd.read_csv('/Users/katychow/Projects/quantified_self/Data/Clean_Header_20190130.csv')
```

```python
df.set_index('dates',inplace=True)
```

```python
df.head(10)
```

```python
# df['asleep_rolling7'] = df['asleep_min'].rolling(7).mean()
df['restless_rolling7'] = df['restless_min'].rolling(7).mean()

df['awake_rolling7'] = df['awake_min'].rolling(7).mean()
```

```python
df.fillna(0,inplace=True)
```

```python
df['dayofweek']=df.index
df['dayofweek']=[ pd.to_datetime(i).weekday() for i in df['dayofweek']]
```

```python
train = df[:90]
test = df[90:]
```

```python
train.shape, test.shape
```

```python
features = ['asleep_min', 'awake_min', 'restless_min', 'restless_rolling7', 'awake_rolling7']
train = train[features]
test = test[features]

train.shape, test.shape
```

```python
### how to predict 1 variable based on 1 dataset
```

```python
from statsmodels.tsa.stattools import adfuller
adfuller(train['asleep_min'])
# definitely not stationary
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(train['asleep_min'], model='additive', freq=6)

fig = plt.figure()  
fig = result.plot()  
fig.set_size_inches(15, 12)
```

```python
plot_acf(train['asleep_min'],lags=24) # day 7 and day 12 spikes
```

```python
ar_eff = ARIMA(endog = train['asleep_min'], order = (1,0,0))
model = ar_eff.fit()
preds = model.predict(start = 90, end = 120)
mean_squared_error(test['asleep_min'],preds)
```

```python
sarima = SARIMAX( endog = train['asleep_min'],
                  order = (6,0,2),              # (p, d, q) #7,0,2 & 7,0,2,7 worked well too 
                  seasonal_order= (6,0,2,6)) # (P, D, Q, S)

# Fit SARIMA model.
model = sarima.fit()

# Generate predictions based on test set.
preds = model.predict(start = 90, end = 120)

print(mean_squared_error(test['asleep_min'], preds))
```

```python
mean_absolute_error(test['asleep_min'],preds)
```

```python
preds = pd.DataFrame(preds)
preds.columns = ['pred_asleep_min']
```

```python
preds.set_index(test.index,inplace = True)
```

```python
preds.plot(color = 'pink')
test['asleep_min'].plot(color='lightsteelblue')
```
